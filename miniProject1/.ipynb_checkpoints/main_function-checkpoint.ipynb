```python
from rank import *
def toUpper(value):
    if isinstance(value,str):
        return value.upper()
    elif isinstance(value,list):
        return [item.upper() for item in value]
    elif isinstance(value,set):
        lst = [item.upper() for item in value]
        return set(lst)
```
